```python
from textblob import TextBlob
```

```python
from newspaper import Article
```

```python
url = 'https://en.wikipedia.org/wiki/COVID-19'
```

```python
article = Article(url)
```

```python
article.download()
```

```python
article.parse()
article.nlp()
```

```python
text = article.text
print(text)
```

```python
text = article.summary
print(text)
```

```python
blob = TextBlob(text)
```

```python
sentiment = blob.sentiment.polarity # -1 to 1
```

```python
print(sentiment)
```

```python
# This article is leaning more towards a negative polarity.
```
